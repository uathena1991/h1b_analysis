```python
def data_cleansing(raw_data, column_idx):
    """
    just in case of: Case issue, capitalize all words
    return:  UPPER cases data, as np.array
    """
    res = np.array([[x.upper() for x in y] for y in raw_data])
    return res
```

```python
def top_n_freq_items(certified_data, key_col_idx, num_tops = 10):
    """
    :params: certified_data: np.array, input data (with 'CERTIFIED' status)
    :params: key_col_idx: int, states column idx
    :params: visa_col_idx: int, visa column idx
    :params: num_states: int, top num_states frequent states
    
    returns:
        res: [[OCCUPATION/STATES, NUMBER_CERTIFIED_APPLICATIONS, PERCENTAGE]]
    """
    ## filter all data by visa_col_idx
    
    ## using counter
    ct = Counter()
    for st in certified_data[:, key_col_idx]:
        ct[st] += 1
    del ct['']
    total_sum = len(certified_data)
    res = [[x[0], x[1], round(x[1]/total_sum, 3)] for x in ct.most_common(num_tops)] # top_occupations, number_certified_applications, percentage
    return res, ct
    ## using numpy.unique -- slower than Counter.... 
#     unique, counts = np.unique(data[:,key_col_idx], return_counts=True)
#     return unique[np.argsort(counts)][::-1][:num_states]

# t1 = time.time()
# top_n_freq_items(tmp_data, 11, 10)
# print(time.time() - t1)
```

```python
def txt_writer(data, filename, path, headers):
    """
    write data to a file
    """
    try:
        with open(os.path.join(path, filename), 'w') as file:
            ## write headers
            file.write(headers)
            file.write("\n")
            ## write data
            for row in data:
                file.write("%s;%d;%1.1f%%\n" %(row[0], row[1], row[2]*100))
        file.close()
    except ValueError:
        print("ERROR")
        return False
    return True
    
```

```python
def main(filename, parent_path, 
         visa_key = "STATUS", state_key = "LCA_CASE_EMPLOYER_STATE", occ_key = "LCA_CASE_SOC_NAME",
         cell_delimiter =  ";", header_idx = True, 
         certified_str = 'CERTIFIED', num_top = 10,
        out_occu_fn = "top_10_occupations.txt", out_state_fn = 'top_10_states.txt', 
         input_path = 'input', out_path = 'output',
        headers_states = "TOP_STATES;NUMBER_CERTIFIED_APPLICATIONS;PERCENTAGE", 
        headers_occ = "TOP_OCCUPATIONS;NUMBER_CERTIFIED_APPLICATIONS;PERCENTAGE"):
    """
    main function to call all modules
    """
    ## load data
    raw_data, headers, trouble_data = compile_csv(filename, os.path.join(parent_path, input_path))
    ## find visa column idx:
    visa_column = headers.index(visa_key)
    state_column = headers.index(state_key)
    occ_column = headers.index(occ_key)

    ## data cleansing (like upper case)
    ## get certified data
    certified_data = raw_data[raw_data[:,visa_column] == certified_str]
    top_states, all_states = top_n_freq_items(certified_data, state_column, num_top)
    top_occ, all_occ = top_n_freq_items(certified_data, occ_column, num_top)
    ## write to txt files
    stat_states = txt_writer(top_states, out_state_fn, os.path.join(parent_path, out_path), headers_states)
    stat_occ = txt_writer(top_occ, out_occu_fn, os.path.join(parent_path, out_path), headers_occ)
    if stat_states and stat_occ: 
        return True
    else:
        return False
main('H1B_FY_2016.csv', parent_path, 'CASE_STATUS', 'EMPLOYER_STATE', 'SOC_NAME')
```
